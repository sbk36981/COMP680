```python
import pandas as pd
import numpy as np
import matplotlib as plt
from urllib.parse import urlparse
from datetime import datetime, date, time
```

```python
fname = 'example_data/access.log'
df = pd.read_csv(fname, sep=' ',header=None)
df.info()
```

Parse URL into parameters and endpoint

```python
def splitURL(row):
    urlobj = urlparse(row[7])
    row['url_netloc'] = urlobj.netloc
    row['url_params'] = urlobj.params
    row['url_query'] = urlobj.query
    row['url_fragment'] = urlobj.fragment
    return row
df = df.apply(splitURL, axis=1)
```

```python
df.head()
```

```python
def dateparser(row):
    datestring = row[3]
    row[3] = datetime.strptime(datestring, "[%d/%b/%Y:%H:%M:%S]")
    return row
df = df.apply(dateparser, axis=1)
df.head()
```
